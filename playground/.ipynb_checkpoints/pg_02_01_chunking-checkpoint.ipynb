# Simple character based chunking

```python
def sliding_window(seq: list, chunk_size: int, step: int) -> list[dict]:
    if chunk_size <= 0 or step <= 0:
        raise ValueError("Must be greater than 0!")
    result = []
    for i in range(0, len(seq), step):
        chunk = seq[i:i+chunk_size]
        result.append({"start": i, "chunk": chunk})
        if i + chunk_size > len(seq):
            break
    return result
```

```python
result = sliding_window(seq=data[45]["content"], chunk_size=2000, step=1000)
```

```python
result[0:2]
```

```python
def chunk_everything(documents: list[dict]) -> list[dict]:
    all_chunks = []
    
    for doc in documents:
        doc_copy = doc.copy()
        content = doc_copy.pop("content")
        chunks = sliding_window(seq=content, chunk_size=2000, step=1000)
        for chunk in chunks:
            chunk.update(doc_copy) # Adds the remaining data from doc_copy to chunk (title', 'description', 'filename')
        all_chunks.extend(chunks)
    return all_chunks
```

```python
all_chunks = chunk_everything(documents=data)
all_chunks[100]
```

# Section based chunking

```python
import re
from typing import List, Dict

def split_markdown_by_heading(seq: str, level: int = 1) -> List[Dict[str, str]]:
    """
    Split markdown text into chunks based on a specific heading level.

    Each chunk includes:
      - 'start': starting character index of the chunk
      - 'chunk': the full text of the chunk (including heading)

    Args:
        seq (str): Markdown content
        level (int): Heading level (1 = '#', 2 = '##', ...)

    Returns:
        List[Dict]: [{'start': int, 'chunk': str}, ...]
    """
    if level < 1:
        raise ValueError("Heading level must be >= 1")

    pattern = re.compile(rf"^(#{level})\s+.*", re.MULTILINE)
    matches = list(pattern.finditer(seq))

    chunks = []

    for i, match in enumerate(matches):
        start = match.start()
        end = matches[i + 1].start() if i + 1 < len(matches) else len(seq)

        chunk_text = seq[start:end].strip()

        chunks.append({
            "start": start,
            "chunk": chunk_text
        })

    return chunks
```
